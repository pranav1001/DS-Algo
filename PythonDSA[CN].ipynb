```
def isSorted(a):
  l= len(a)
  if l== 0 or l==1:
    return True
  
  if a[0]>a[1]:
    return False

  smallerList= a[1:]
  isSmallerListSorted = isSorted(smallerList)
  if isSmallerListSorted:
    return True
  else:
    return False
  
```

```
''' Optimized code'''
def isSorted2(a, si):
  l= len(a)
  if si== l-1 or si==l:
    return True
  if a[si]> a[si+1]:
    return False
  isSmallerPartSorted= isSorted2(a,si+1)
  return isSmallerPartSorted
```

```
a= [1,2,3,4,5]
isSorted2(a,0)
```

```
'''Find the First index of the element'''
def firstIndex(a, x):
  l= len(a)
  if l== 0:
    return -1
  
  if a[0]==x:
    return 0
  smallerList= a[1:]
  smallerListOutput= firstIndex(smallerList, x)
  if smallerListOutput == -1:
    return -1
  else:
    return smallerListOutput + 1
```
